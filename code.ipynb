```python
## now we convert the string of list to list
## we used a model for this purpose(literal_eval from ast)
```

```python
import ast
def converter(text):
    L = []
    for i in ast.literal_eval(text):
        L.append(i['name']) 
    return L
```

```python
movies['genres']=movies['genres'].apply(converter)
```

```python
movies.genres.head(1)
```

```python
movies['keywords']=movies['keywords'].apply(converter)
```

```python
## we take only three name from cast and crew
```

```python
def converter3(text):
    L = []
    counter=0
    for i in ast.literal_eval(text):
        if counter<3:
            L.append(i['name']) 
        counter=counter+1
    return L
```

```python
movies['crew']=movies['crew'].apply(converter3)
movies['cast']=movies['cast'].apply(converter3)
```

```python
movies.head(1)
```

```python
## now we remove the spaces
```

```python
def collapse(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```

```python
movies['genres']=movies['genres'].apply(collapse)
movies['cast']=movies['cast'].apply(collapse)
movies['crew']=movies['crew'].apply(collapse)
movies['keywords']=movies['keywords'].apply(collapse)
```

```python
movies.head()
```

```python
movies.isnull().sum()
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies.overview.head(1).values
```

```python
##movies.overview.dtypes
```

```python
## Now we have to concate the columns to make one column
movies['tag']=movies['overview']+movies['cast']+movies['crew']+movies['genres']+movies['keywords']
```

```python
final_movies=movies.drop(['cast','crew','genres','keywords','overview'],axis=1)
```

```python
final_movies
```

```python
final_movies.isnull().sum()
```

```python
final_movies['tag']=final_movies['tag'].apply(lambda x : " ".join(x))
```

```python
final_movies.head(1).values
```

```python
def lower(text):
        return text.lower()
final_movies["tag"]=final_movies["tag"].apply(lower)
```

```python
## now we use nltk library to clean data in form of lower,stopwards etc
```

```python
import nltk
import re
```

```python
from nltk.corpus import stopwords
nltk.download("stopwords")
```

```python
## applying stemming (removiing verb words ex love,loves,loved,loving  ,these word have same meaning)
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
        
    return " ".join(y)
```

```python
final_movies["tag"]=final_movies["tag"].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words='english')
```

```python
vector=cv.fit_transform(final_movies['tag']).toarray()
```

```python
vector
```

```python
vector.shape
```

```python
## five thousand most common words
```

```python
cv.get_feature_names()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vector)
```

```python
final_movies[final_movies['title']=='Avatar'].index[0]
```

```python
def recommend(movie):
    index=final_movies[final_movies['title']==movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(final_movies.iloc[i[0]].title)
```

```python
recommend('Spectre')
```
